```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sachin = [500, 700, 1100, 1500, 1800]
sehwag = [0, 200, 900, 1400, 1600]
kohli  = [0, 0, 500, 800, 1100]
years = [1990, 1992, 1994, 1996, 1998]
x=np.arange(len(years))
width=0.30
# for i in range (len(years)):
#     plt.text(i,years[i]-60,str(years[i]),ha="center")
plt.xticks(x,years)
plt.bar(x-width, sachin, width=width, label="sachin")
plt.bar(x,sehwag,width=width,label="sehwag")
plt.bar(x+width,kohli,width=width, label="kohli")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()
```
