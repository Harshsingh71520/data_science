```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y1 = [i * 2 for i in x]
y2 = [i ** 2 for i in x]
y3= [i** 0.5 for i in x]
y4=[10-i for i in x]
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title("Double of x")
plt.xlabel("Number")
plt.ylabel("Double of No.")

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title("Square of x")
plt.xlabel("Number")
plt.ylabel("  Square of No.")
plt.tight_layout()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("Double of x")
plt.xlabel("Number")
plt.ylabel("Double of No.")
plt.grid(True)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("Square of x")
plt.xlabel("Number")
plt.ylabel("  Square of No.")
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("Sqrt of x")
plt.xlabel("Number")
plt.ylabel("Sqrt of No.")
plt.grid(True)
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("difference of x")
plt.xlabel("Number")
plt.ylabel("Diff of No.")
plt.tight_layout()
plt.grid(True)
```
