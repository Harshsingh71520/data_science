```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("fivethirtyeight")
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
plt.pie(runs, labels=labels)
plt.title("carrer run of Indian batmen's")
# plt.pie(runs, labels=labels)
```

```python
plt.pie(
    runs,
    labels=labels,
    shadow= True,
    startangle=50,
    # colors=colors,
    wedgeprops={"edgecolor":"black","linewidth":2 ,"linestyle":"--"}
)
```

```python
 explode=[0.1,0.02,0.15,0.07,]
plt.pie(runs, labels=labels,explode=explode,autopct="%1.1f%%")
plt.title("Graph")
```
