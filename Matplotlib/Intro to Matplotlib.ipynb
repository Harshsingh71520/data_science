```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3],[4,5,6])
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
 
plt.plot(years, runs) 
plt.show()
```

```python
plt.plot(years, runs)
plt.xlabel("Years")
plt.ylabel("runs")
plt.title("Sachin Tendulkar's yearly Runs")
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
Sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label="Virat Kholi")
plt.plot(years, Sehwag, label="Sehwag")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Runs Scored")
plt.legend()
```

```python
plt.plot(years, kohli,"ro--", label="Virat Kholi")
plt.plot(years, Sehwag,"g+:", label="Sehwag")
plt.legend()
plt.grid(True)
```

```python
Kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
Sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
Rohit = [0, 500, 800, 1100, 1300,1500, 1700, 1600, 1400,1500]
plt.plot(years, Kohli,"b*:", label="Virat Kohli")
plt.plot(years, Sehwag,"g+:", label ="Sehwag")
plt.plot(years, Rohit,"ro--", label="Rohit")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()
plt.grid(True)
```
